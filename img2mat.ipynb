```python

import h5py
import cv2
import torch
import numpy as np
import os
import scipy
import sys

new_path = r"D:\workspace\python_ws\pose-master"
sys.path.append(new_path)

from load_dataset import SkeletonDataset
```

```python
def to_mat(save_path, dict, i):
    '''
    提取需要的pressure map, skeleton, trans, gender 保存为.mat格式
    save_path------train\test---name--- 1.mat, 2.mat, ...
    '''
    # print(f'coverting file {self.name} to .mat')
    # root_dir = os.path.join(save_path)
    os.makedirs(save_path, exist_ok=True)
    skeleton_annotations_path = os.path.join(save_path,  f'{i}.mat')
    '''
     MAT 文件（MATLAB 格式文件）中没有直接对应的布尔数据类型，
     因此在保存时将布尔数组视为无符号 8 位整数（uint8）来处理
    '''
    scipy.io.savemat(skeleton_annotations_path, {'image':dict['image'].numpy(),
                                                 'skeleton':dict['skeleton'].numpy(),
                                                 'trans':dict['trans'].numpy(),
                                                 'gender':dict['gender'].numpy()
                                                    })
```

```python
# train数据预处理并保存为.mat
root_dir = r'D:\workspace\python_ws\pose-master\dataset\train'
for curr_path,folders,files in os.walk(root_dir):
    # skeletons = []
    # images = []
    # transs = []
    # genders = []
    # 跳过根目录
    if curr_path == root_dir:
        continue
    print("当前目录：",curr_path)
    # print("当前目录：",os.path.basename(curr_path))
    # print("当前目录下的文件夹：",folders)
    # print("当前目录下的文件：",files)
    path_name = str(os.path.basename(curr_path))
    dataset = SkeletonDataset(curr_path, transform=True)
    # batch_size = 64
    data_loader = torch.utils.data.DataLoader(dataset, batch_size=None, shuffle=False)
    for i, batch in enumerate(data_loader):
        # print(batch)
        print(f"saving... [{i}.mat]",end="\r")
        to_mat(save_path=fr'F:\pose_master_dataset_mat\train\{path_name}',dict=batch,i=i)
        # if i >= 10:
        #     break
    # break
    print(f"saving data {path_name} compeleted!")
```

```python
# test数据集预处理并保存为.mat
root_dir = r'D:\workspace\python_ws\pose-master\dataset\test'
for curr_path,folders,files in os.walk(root_dir):

    # 跳过根目录
    if curr_path == root_dir:
        continue
    print("当前目录：",curr_path)

    path_name = str(os.path.basename(curr_path))
    dataset = SkeletonDataset(curr_path, transform=True)

    data_loader = torch.utils.data.DataLoader(dataset, batch_size=None, shuffle=False)
    for i, batch in enumerate(data_loader):
        # print(batch)
        print(f"saving... [{i}.mat]",end="\r")
        to_mat(save_path=fr'F:\pose_master_dataset_mat\test\{path_name}',dict=batch,i=i)

    print(f"saving data [{path_name}] compeleted!")
```
